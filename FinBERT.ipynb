### Finbert Model 

```python
# df = pd.read_csv("reuters_headlines.csv", nrows=50)
# df_array = np.array(df)
df_list = ['We made a huge loss of 1 million dollars and company is about to close'] 
tokenizer = AutoTokenizer.from_pretrained("ProsusAI/finbert")
model = AutoModelForSequenceClassification.from_pretrained("ProsusAI/finbert")
inputs = tokenizer(df_list, padding = True, truncation = True, return_tensors='pt') #tokenize text to be sent to model
outputs = model(**inputs)
predictions = torch.nn.functional.softmax(outputs.logits, dim=-1)
model.config.id2label
positive = predictions[:, 0].tolist()
negative = predictions[:, 1].tolist()
neutral = predictions[:, 2].tolist()
table = {'Headline':df_list, "Positive":positive, "Negative":negative, "Neutral":neutral}      
df2 = pd.DataFrame(table, columns = ["Headline", "Positive", "Negative", "Neutral"])
```

### Predicting on datatset

```python
df_final=pd.DataFrame()
```

```python
import os

# specify the path to the root directory
root_dir = 'FMP'

# loop through all directories in the root directory
for dir_name in os.listdir(root_dir):
    # check if the current item is a directory
    if os.path.isdir(os.path.join(root_dir, dir_name)):
        for year in os.listdir(os.path.join(root_dir, dir_name)):
            # exclude .DS_Store files from the list of directories
            if year != '.DS_Store':
                content=""
                for file in os.listdir(os.path.join(root_dir, dir_name, year)):
                    # exclude .DS_Store files from the list of files
                    if file != '.DS_Store':
                        file_path = os.path.join(root_dir, dir_name, year, file)  # Update with your actual file path
                        with open(file_path, 'r') as file:
                            data = json.load(file)
                        content+=data['content']
                df_list=[content]
                inputs = tokenizer(df_list, padding=True, truncation=True, return_tensors='pt') #tokenize text to be sent to model
                outputs = model(**inputs)
                predictions = torch.nn.functional.softmax(outputs.logits, dim=-1)
                model.config.id2label
                positive = predictions[:, 0].tolist()
                negative = predictions[:, 1].tolist()
                neutral = predictions[:, 2].tolist()
                table = {'company_name': dir_name, 'year': year, "Positive": positive, "Negative": negative, "Neutral": neutral}      
                df_final = df_final.append(table, ignore_index=True)
    print(dir_name)
```
